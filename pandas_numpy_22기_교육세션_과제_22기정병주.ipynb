#### birthDate 컬럼으로 age 컬럼 추가하기

##### 1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
data['birthDate'] = pd.to_datetime(data['birthDate'])
data.info()
data.head()
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
import datetime
def age_cal(x):
    difference = datetime.date.today().year - x.year
    return difference

data['age'] = data['birthDate'].apply(age_cal)
data['age'].head()
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
data['age'].isna().sum()
data['age'] = data['age'].fillna(data['age'].dropna().mean())
data['age'].isna().sum()
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
def height_cm(x):
    lst = str(x).split('-')
    return int(lst[0]) * 30.48 + int(lst[1]) * 2.54
data['heightCm'] = data['height'].apply(height_cm)
data['heightCm'].head()
```
